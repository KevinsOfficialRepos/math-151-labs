### Question 3

#### 3a

```python
# Enter your code here
from sympy import simplify


t = sp.symbols('t')
x = -2*t/(1+t**2)
y = (t**2-1)/(1+t**2)
simple_t = sp.simplify(x/(y-1))
print(simplify(simple_t - t) == 0)
```

#### 3b

```python
# Enter your code here
dx_dt = sp.diff(x, t)
dy_dt = sp.diff(y, t)
horizontal_tangent = sp.solve(dy_dt, t)
horizontal_points = [num for num in horizontal_tangent if dx_dt.subs(t, num) != 0]

vertical_tangent = sp.solve(dx_dt, t)
vertical_points = [num for num in vertical_tangent if dy_dt.subs(t, num) != 0]

print("dx_dt:", dx_dt, "\ndy_dt:", dy_dt)
print("Horizontal tangent points(x,y):", [(x.subs(t, num), y.subs(t, num)) for num in horizontal_points])
print("Vertical tangent points(x,y):", [(x.subs(t, num), y.subs(t, num)) for num in vertical_points])
```
