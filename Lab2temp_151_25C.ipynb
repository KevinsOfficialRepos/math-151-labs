```python
#start code here
x = sp.symbols('x')
f = (sp.sin(2*x))/(x - sp.pi.evalf())
print(f)
lower = [3.1, 3.14, 3.141, 3.1415]
upper = [3.2, 3.15, 3.142, 3.1416]
for i in range(4):
    print(f"f({lower[i]}) = {f.subs(x, lower[i])}")
    print(f"f({upper[i]}) = {f.subs(x, upper[i])}")
```

## Question 3

#### 3a

```python
#start code here
c = sp.symbols('c', positive=True)
t = sp.symbols('t')
x = 2*c*t - 4*t**3
y = -c*t**2 + 3*t**4
p1 = plot_parametric((x.subs(c,2), y.subs(c,2)), (t, -2.5, 2.5), show=False, line_color='blue')
p2 = plot_parametric((x.subs(c,5), y.subs(c,5)), (t, -2.5, 2.5), show=False, line_color='red')
p3 = plot_parametric((x.subs(c,8), y.subs(c,8)), (t, -2.5, 2.5), show=False, line_color='green')
p1.extend(p2)
p1.extend(p3)
p1.title = "Swallowtail Catastrophe Curves for c=2,5,8"
p1.xlabel = "x"
p1.ylabel = "y"
p1.xlim = (-20, 20)
p1.ylim = (-10, 25)
p1.show()
```

#### 3b

```python
#start code here
c = sp.symbols('c', positive=True)
t = sp.symbols('t')
x = 2*c*t - 4*t**3
y = -c*t**2 + 3*t**4
t_val = sp.solve(x, t)
print(t_val)
y_vals = [y.subs(t, t_val[0]), y.subs(t, t_val[1]), y.subs(t, t_val[2])]
print(f"The y-coordinates where the curves intersect the y-axis are: {y_vals}")
```
